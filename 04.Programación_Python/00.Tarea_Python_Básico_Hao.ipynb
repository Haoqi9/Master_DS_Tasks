# **Tarea Juego de la Adivinanza de Número - Hao**

## SET-UP

```python
import funciones_juego as fj
import openpyxl
import os

# Parámetros modificables del juego:
# No soporta números negativos
NUM_MIN_ADIVINDAR = 1
NUM_MAX_ADIVINAR = 1000
RUTA_EXCEL = r'C:\Users\34685\Desktop\Máster_UCM\1.Módulos\4.Programación_Python\A.Python_Básico\4.Tarea'
NOMBRE_ARCHIVO_EXCEL = 'Tarea_Estadisticas.xlsx'
```

## Juego 

```python
# Mensaje inicial del juego.
print('🙌🏼Bienvenido al juego de la adivinanza de número🙌🏼\n')
# print()
print('Aquí están las opciones disponibles. ¡Elije una!')

# Definir el la ruta donde se va a crear/cargar el excel.
os.chdir(RUTA_EXCEL)

# Crear o cargar archivo excel.
if not os.path.exists(RUTA_EXCEL + '\\' + NOMBRE_ARCHIVO_EXCEL):
    # Crear una hoja predefinida 'Total' que contenga las estadísticas agregadas.
    workbook = openpyxl.Workbook()
    workbook.active.title = 'total'
    fj.crear_estructura_hoja_inicial(workbook, 'total')
else:
    workbook = openpyxl.load_workbook(NOMBRE_ARCHIVO_EXCEL)

# Bucle principal del juego.
# Se detiene cuando el jugador así lo indique, la opción salir es elegida, 
# o cuando se quiera visualizar las estadisticas.
opcion_menu = 1
while opcion_menu in [1, 2]:
    
    # Menú principal.
    fj.mostrar_menu()
    opcion_menu = fj.validar_num(1,4)
    
    if opcion_menu in [1, 2]:
        
        # Al parecer los nombres de las hojas de excel son insensibles a mayúsculas y minúsculas.
        print('\n ⚠️ ¡Importante: el juego no distingue entre mayúsculas y minúsculas! ⚠️')
        print("💡 Ejemplo: nombres tanto 'Hao' como 'hAo' serán tratados como 'hao'.")
        print('➡️ Aquí tienes una lista de los nombres de jugadores 2 anteriores:')
        print('➡️', workbook.sheetnames[1:], '\n')
        
        nombre_jugador = input('🙋‍♂️ Introduzca el nombre del jugador que adivina (jugador 2): ').lower()
        while nombre_jugador == 'total':
            nombre_jugador = input("El nombre 'total' está reservado. Escoja otro nombre: ").lower()
        
        # Guardar nombres nuevos en una nueva hoja en el excel.
        if nombre_jugador not in workbook.sheetnames:
            workbook.create_sheet(nombre_jugador)
            fj.crear_estructura_hoja_inicial(workbook, nombre_jugador)
        
        print()  # salto de línea entre mensajes.
        # Menú con opciones de dificultad.
        fj.mostrar_sub_menu()
        opcion_dificultad = fj.validar_num(1,3)
        num_intentos = fj.obtener_numero_intentos(opcion_dificultad)
        
        if opcion_menu == 1:
            numero_secreto = fj.obtener_numero_solitario(NUM_MIN_ADIVINDAR, NUM_MAX_ADIVINAR)
            
        elif opcion_menu == 2:
            print('\n👀 ¡Turno del jugador 1 para decidir el número a adivinar!')
            numero_secreto = fj.obtener_numero_2jugadores(NUM_MIN_ADIVINDAR, NUM_MAX_ADIVINAR)

        print(f'\nTurno del jugador 2 para adivinar el número secreto ({NUM_MIN_ADIVINDAR}-{NUM_MAX_ADIVINAR}).')
        print(f'Tienes {num_intentos} intentos en total para adivinarlo. ¡Suerte!💪\n')

        for intento in range(1, num_intentos + 1):
            
            numero_adivinado = fj.obtener_numero_adivinado(NUM_MIN_ADIVINDAR, NUM_MAX_ADIVINAR, intento, num_intentos)
                
            if numero_secreto > numero_adivinado:
                print('¡El número secreto es mayor que el adivinado!\n')
            elif numero_secreto < numero_adivinado:
                print('¡El número secreto es menor que el adivinado!\n')
            elif numero_secreto == numero_adivinado:
                print(f'\n🥳¡Enhorabuena, el número secreto es {numero_adivinado}!🥳')
                break
            
        if numero_secreto != numero_adivinado:
            ganada = False
            print(f'Lástima, se te acabaron los intentos. El número secreto es {numero_secreto}.😞')
        else:
            ganada = True
        
        # Actualizar y guardar los datos de cada partida terminada del jugador en el excel.
        fj.actualizar_estadisticas(workbook, nombre_jugador, opcion_dificultad, ganada)
        workbook.save(NOMBRE_ARCHIVO_EXCEL)
        
    elif opcion_menu == 3:
        print()
        fj.mostrar_menu_estadisticas()
        opcion_visualizacion = fj.validar_num(1, 2)
        
        print("\n👉 Primero debes seleccionar un jugador cuyas estadísticas deseas visualizar.")
        print("💡'total' muestra las estadísticas agregadas de todos los jugadores.")
        print("Aquí tienes una lista con los nombres de todos los jugadores: ")
        print(workbook.sheetnames)
    
        jugador_visualizacion = ''
        while jugador_visualizacion not in workbook.sheetnames:
            jugador_visualizacion = input('Escriba un nombre de la lista: ')
        
        print()
        if opcion_visualizacion == 1:
            fj.visualizar_estadisticas_descriptivas(workbook, jugador_visualizacion)
            
        elif opcion_visualizacion == 2:
            fj.visualizar_estadisticas_graficas(workbook, jugador_visualizacion)
    
    elif opcion_menu == 4:
        pass
    
    # Preguntar al jugador si quiere volver al menú principal.
    if opcion_menu in [1, 2]:
        respuesta = fj.obtener_respuesta()
        if respuesta == 'no':
            print('✨ ¡Gracias por jugar! Espero que hayas disfrutado del juego ✨')
            break
```
