## SET-UP

```python
import funciones_juego as fj
import openpyxl
import os

# ParÃ¡metros modificables del juego:
# No soporta nÃºmeros negativos
NUM_MIN_ADIVINDAR = 1
NUM_MAX_ADIVINAR = 1000
RUTA_EXCEL = r'C:\Users\34685\Desktop\MÃ¡ster_UCM\1.MÃ³dulos\4.ProgramaciÃ³n_Python\A.Python_BÃ¡sico\4.Tarea'
NOMBRE_ARCHIVO_EXCEL = 'Tarea_Estadisticas.xlsx'
```

## Juego 

```python
# Mensaje inicial del juego.
print('ðŸ™ŒðŸ¼Bienvenido al juego de la adivinanza de nÃºmeroðŸ™ŒðŸ¼\n')
# print()
print('AquÃ­ estÃ¡n las opciones disponibles. Â¡Elije una!')

# Definir el la ruta donde se va a crear/cargar el excel.
os.chdir(RUTA_EXCEL)

# Crear o cargar archivo excel.
if not os.path.exists(RUTA_EXCEL + '\\' + NOMBRE_ARCHIVO_EXCEL):
    # Crear una hoja predefinida 'Total' que contenga las estadÃ­sticas agregadas.
    workbook = openpyxl.Workbook()
    workbook.active.title = 'total'
    fj.crear_estructura_hoja_inicial(workbook, 'total')
else:
    workbook = openpyxl.load_workbook(NOMBRE_ARCHIVO_EXCEL)

# Bucle principal del juego.
# Se detiene cuando el jugador asÃ­ lo indique, la opciÃ³n salir es elegida, 
# o cuando se quiera visualizar las estadisticas.
opcion_menu = 1
while opcion_menu in [1, 2]:
    
    # MenÃº principal.
    fj.mostrar_menu()
    opcion_menu = fj.validar_num(1,4)
    
    if opcion_menu in [1, 2]:
        
        # Al parecer los nombres de las hojas de excel son insensibles a mayÃºsculas y minÃºsculas.
        print('\n âš ï¸ Â¡Importante: el juego no distingue entre mayÃºsculas y minÃºsculas! âš ï¸')
        print("ðŸ’¡ Ejemplo: nombres tanto 'Hao' como 'hAo' serÃ¡n tratados como 'hao'.")
        print('âž¡ï¸ AquÃ­ tienes una lista de los nombres de jugadores 2 anteriores:')
        print('âž¡ï¸', workbook.sheetnames[1:], '\n')
        
        nombre_jugador = input('ðŸ™‹â€â™‚ï¸ Introduzca el nombre del jugador que adivina (jugador 2): ').lower()
        while nombre_jugador == 'total':
            nombre_jugador = input("El nombre 'total' estÃ¡ reservado. Escoja otro nombre: ").lower()
        
        # Guardar nombres nuevos en una nueva hoja en el excel.
        if nombre_jugador not in workbook.sheetnames:
            workbook.create_sheet(nombre_jugador)
            fj.crear_estructura_hoja_inicial(workbook, nombre_jugador)
        
        print()  # salto de lÃ­nea entre mensajes.
        # MenÃº con opciones de dificultad.
        fj.mostrar_sub_menu()
        opcion_dificultad = fj.validar_num(1,3)
        num_intentos = fj.obtener_numero_intentos(opcion_dificultad)
        
        if opcion_menu == 1:
            numero_secreto = fj.obtener_numero_solitario(NUM_MIN_ADIVINDAR, NUM_MAX_ADIVINAR)
            
        elif opcion_menu == 2:
            print('\nðŸ‘€ Â¡Turno del jugador 1 para decidir el nÃºmero a adivinar!')
            numero_secreto = fj.obtener_numero_2jugadores(NUM_MIN_ADIVINDAR, NUM_MAX_ADIVINAR)

        print(f'\nTurno del jugador 2 para adivinar el nÃºmero secreto ({NUM_MIN_ADIVINDAR}-{NUM_MAX_ADIVINAR}).')
        print(f'Tienes {num_intentos} intentos en total para adivinarlo. Â¡Suerte!ðŸ’ª\n')

        for intento in range(1, num_intentos + 1):
            
            numero_adivinado = fj.obtener_numero_adivinado(NUM_MIN_ADIVINDAR, NUM_MAX_ADIVINAR, intento, num_intentos)
                
            if numero_secreto > numero_adivinado:
                print('Â¡El nÃºmero secreto es mayor que el adivinado!\n')
            elif numero_secreto < numero_adivinado:
                print('Â¡El nÃºmero secreto es menor que el adivinado!\n')
            elif numero_secreto == numero_adivinado:
                print(f'\nðŸ¥³Â¡Enhorabuena, el nÃºmero secreto es {numero_adivinado}!ðŸ¥³')
                break
            
        if numero_secreto != numero_adivinado:
            ganada = False
            print(f'LÃ¡stima, se te acabaron los intentos. El nÃºmero secreto es {numero_secreto}.ðŸ˜ž')
        else:
            ganada = True
        
        # Actualizar y guardar los datos de cada partida terminada del jugador en el excel.
        fj.actualizar_estadisticas(workbook, nombre_jugador, opcion_dificultad, ganada)
        workbook.save(NOMBRE_ARCHIVO_EXCEL)
        
    elif opcion_menu == 3:
        print()
        fj.mostrar_menu_estadisticas()
        opcion_visualizacion = fj.validar_num(1, 2)
        
        print("\nðŸ‘‰ Primero debes seleccionar un jugador cuyas estadÃ­sticas deseas visualizar.")
        print("ðŸ’¡'total' muestra las estadÃ­sticas agregadas de todos los jugadores.")
        print("AquÃ­ tienes una lista con los nombres de todos los jugadores: ")
        print(workbook.sheetnames)
    
        jugador_visualizacion = ''
        while jugador_visualizacion not in workbook.sheetnames:
            jugador_visualizacion = input('Escriba un nombre de la lista: ')
        
        print()
        if opcion_visualizacion == 1:
            fj.visualizar_estadisticas_descriptivas(workbook, jugador_visualizacion)
            
        elif opcion_visualizacion == 2:
            fj.visualizar_estadisticas_graficas(workbook, jugador_visualizacion)
    
    elif opcion_menu == 4:
        pass
    
    # Preguntar al jugador si quiere volver al menÃº principal.
    if opcion_menu in [1, 2]:
        respuesta = fj.obtener_respuesta()
        if respuesta == 'no':
            print('âœ¨ Â¡Gracias por jugar! Espero que hayas disfrutado del juego âœ¨')
            break
```
